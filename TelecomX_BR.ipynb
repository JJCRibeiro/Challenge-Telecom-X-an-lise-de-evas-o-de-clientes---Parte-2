**3. Balanceamento de Classes:**
Considerando o desequilíbrio (73,5% vs 26,5%), podemos aplicar técnicas como:

````python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

# 1. Pré-processamento necessário
# Garantir que não há valores ausentes e churn é numérico
df_codificado['churn'] = pd.to_numeric(df_codificado['churn'], errors='coerce')
df_codificado = df_codificado.dropna(subset=['churn'])

# 2. Separar features e target
X = df_codificado.drop('churn', axis=1)
y = df_codificado['churn']

# 3. Verificar tipos de dados - todas features devem ser numéricas
print("\nTipos de dados das features:")
print(X.dtypes)

# 4. Aplicar SMOTE (oversampling)
try:
    smote = SMOTE(random_state=42)
    X_res_smote, y_res_smote = smote.fit_resample(X, y)
    print("\nSMOTE aplicado com sucesso!")
    print(f"Shape antes: {X.shape}, Shape após SMOTE: {X_res_smote.shape}")
    print("Distribuição após SMOTE:", pd.Series(y_res_smote).value_counts())
except Exception as e:
    print("\nErro ao aplicar SMOTE:", e)

# 5. Aplicar Undersampling
try:
    under = RandomUnderSampler(random_state=42)
    X_res_under, y_res_under = under.fit_resample(X, y)
    print("\nUndersampling aplicado com sucesso!")
    print(f"Shape antes: {X.shape}, Shape após undersampling: {X_res_under.shape}")
    print("Distribuição após undersampling:", pd.Series(y_res_under).value_counts())
except Exception as e:
    print("\nErro ao aplicar undersampling:", e)
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import StandardScaler
import numpy as np # Import numpy for isnan check

# Pipeline with normalization and model
lr_pipe = Pipeline([
    ('scaler', StandardScaler()),  # Normalization important for Logistic Regression
    ('classifier', LogisticRegression(
        random_state=42,
        class_weight='balanced',  # To handle imbalance
        max_iter=1000
    ))
])

# Training
# Explicitly check for NaNs before fitting
if np.isnan(X_train).sum().sum() > 0 or np.isnan(y_train).sum() > 0:
    print("Error: NaN values found in X_train or y_train. Cannot fit the model.")
    if np.isnan(X_train).sum().sum() > 0:
        print("Total NaNs in X_train:", np.isnan(X_train).sum().sum())
        # Optional: print columns with NaNs
        # print(X_train.columns[X_train.isnull().any()].tolist())
    if np.isnan(y_train).sum() > 0:
        print("Total NaNs in y_train:", np.isnan(y_train).sum())
    lr_fitted = False # Set a flag to indicate if fitting was successful
else:
    # Assert that there are no NaNs
    assert np.isnan(X_train).sum().sum() == 0, "X_train contains NaN values"
    assert np.isnan(y_train).sum() == 0, "y_train contains NaN values"

    lr_pipe.fit(X_train, y_train)
    lr_fitted = True # Set a flag to indicate if fitting was successful
    print("Logistic Regression model fitted successfully.")

    # Evaluation
    y_pred_lr = lr_pipe.predict(X_test)
    y_proba_lr = lr_pipe.predict_proba(X_test)[:, 1]

    print("Logistic Regression - Classification Report:")
    print(classification_report(y_test, y_pred_lr))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_proba_lr):.4f}")
````

# **Avaliação dos Modelos Preditivos:**

Vamos avaliar sistematicamente os dois modelos criados (Regressão Logística e Random Forest) utilizando as métricas solicitadas e fazer uma análise comparativa crítica.

**1. Métricas de Avaliação:**

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

def evaluate_model(model, X_test, y_test, model_name):
    # Previsões
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    # Cálculo das métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Exibição das métricas
    print(f"\n{'='*50}")
    print(f"{model_name.upper()} - Métricas de Avaliação")
    print(f"{'='*50}")
    print(f"Acurácia: {accuracy:.4f}")
    print(f"Precisão: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")

    # Matriz de Confusão
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title(f'Matriz de Confusão - {model_name}')
    plt.show()

    return {'accuracy': accuracy, 'precision': precision, 'recall': recall, 'f1': f1}
````

**2. Análise Comparativa dos Modelos**

````python
import pandas as pd

# Avaliar Regressão Logística
# Ensure lr_pipe is fitted before evaluating
if 'lr_fitted' in locals() and lr_fitted:
    lr_metrics = evaluate_model(lr_pipe, X_test, y_test, "Regressão Logística")
else:
    print("Logistic Regression model was not fitted successfully. Cannot evaluate.")
    lr_metrics = None # Assign None if model wasn't fitted

# Avaliar Random Forest
rf_metrics = evaluate_model(rf_model, X_test, y_test, "Random Forest")

# Create comparison DataFrame only if Logistic Regression was fitted
if lr_metrics is not None:
    comparison = pd.DataFrame({
        'Regressão Logística': lr_metrics,
        'Random Forest': rf_metrics
    }).T

    print("\nComparação dos Modelos:")
    print(comparison)
else:
    print("\nComparison table cannot be created as Logistic Regression model was not evaluated.")
````

**Análise Crítica:**

Desempenho Geral:

Random Forest tende a apresentar melhor desempenho em F1-Score (média harmônica entre precisão e recall)

Regressão Logística pode ter melhor precisão em alguns casos

Matriz de Confusão:

Analisar falsos positivos e falsos negativos

Random Forest geralmente tem melhor balanceamento entre os erros

**3.Overfitting/Underfitting:**

````python
# Verificar desempenho em treino vs teste
print("\nDesempenho em Treino vs Teste:")

def check_overfitting(model, X_train, y_train, X_test, y_test, name):
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    print(f"{name}:")
    print(f"  Treino: {train_score:.4f} | Teste: {test_score:.4f}")
    print(f"  Diferença: {abs(train_score - test_score):.4f}")
    return train_score, test_score

# Check if lr_pipe was fitted before evaluating
if 'lr_fitted' in locals() and lr_fitted:
    lr_train, lr_test = check_overfitting(lr_pipe, X_train, y_train, X_test, y_test, "Reg. Logística")
else:
    print("Reg. Logística: Model not fitted. Skipping overfitting check.")

# Evaluate Random Forest
rf_train, rf_test = check_overfitting(rf_model, X_train, y_train, X_test, y_test, "Random Forest")
````

Overfitting (se performance no treino >> teste):

Possíveis causas: modelo muito complexo

Soluções: reduzir profundidade das árvores (para RF), aumentar regularização (para RL)

Underfitting (se performance baixa em ambos):

Possíveis causas: modelo muito simples

Soluções: aumentar complexidade, adicionar features
